## Sidebar
The sidebar contains panels that represent different views of the files in your workbench.  You can collapse or expand the sidebar by clicking the toggle button (hotkey: *Cmd-Ctrl-B*) in the top right corner of the workbench, below the navigation bar.  

You can re-arrange panels within the sidebar by dragging the panel's title bar and dropping it above or below other panels, as shown in the following video.

<video controls preload autoplay width="640" height="480" src="http://ibmjstart.com/docdepot/unshackled/kawb/kawb_sidebar.mp4" style="margin-top: 20px"></video>

### Recent Notebooks Panel
Lists the most recently accessed notebooks.  This panel is always shown.

### Recent Data Panel
Lists the most recently accessed data files.  This panel is always shown.

<div class="alert alert-block alert-info" style="margin-top: 20px">**Note:** Currently, Cognitive Class Labs treats any file that is *not* an IPython notebook as a data file.</div>

### Recently Active Panel
When you click on a the name of a notebook, Cognitive Class Labs opens the notebook in a new view.  A kernel process is started on the server and associated with the notebook.  This kernel process is what executes the code cells and stores objects in memory.  For as long as the kernel process is running, the notebook is said to be 'active' or 'running'.

You can create a panel that shows the most recently accessed notebooks with active kernels.  Simply drag the 'Active' text for an active notebook onto the sidebar to create this panel.

## Organize Files
Cognitive Class Labs allows you to tag individual files within your workbench environment.  You can also create your own custom tag panels, which allows you to organize your files by tag. 

### Tag a File
Tags for an individual file are displayed as a **comma-separated** string under the file name.  

You can double-click the file's tags to edit them.

You can also add a tag to a file by dragging the file and dropping it onto an existing tag panel.

### Tag Panels
List most recently accessed files with the given tag.  To create a custom tag panel, drag a tag and drop it onto the sidebar.

### Recent "untagged" Panel
Files that have no tags display as "untagged".  You can create a panel that lists recently accessed "untagged" files.  Simply drag the "untagged" text and drop it somewhere within the sidebar, in the same way you would to create a custom tag panel.

### Close Panels
You can close custom panels using the drop-down menu in the panel's title bar.  Click on the menu icon located to the right of the panel title and select "Close" from panel's the drop-down menu.

## Manage Files

### Create a New Notebook
To create a new IPython notebook, click the new notebook button located in the top navigation bar (upper right corner of the workbench)

When you create a new notebook, you will see the notebook with the name, "UntitledN.ipynb", in the **Recent Notebooks** sidebar panel.

<div class="alert alert-block alert-info" style="margin-top: 20px">**Note:** New notebooks are named "UntitledN.ipynb", where N is a digit [0-9] that gets incremented automatically.
</div>

You can perform the following actions on files in your workbench.  Click the **>** icon to the left of the file name to expand the section that contains links for each action.

### Rename a File
Change the base name of the file.

### Download a File
Download the file to your computer in compressed archive format.

### Clone a File
Create a copy (or clone) the file.  This creates a new file in your workbench.

### Delete a File
Delete the file from your workbench environment.

### Stop an Active Notebook
Stop the running IPython kernel process associated with the notebook.

### Share/Unshare a File
Create or remove a share link to the file.  Cognitive Class Labs users can obtain a copy of the file if they enter its link in their workbench.
